```python
#using pandas import titanic dataset

import pandas as pd 

Titanic_Dataset = pd.read_csv('Titanic-Dataset.csv')

Titanic_Dataset
```

# 2.Handle missing values using mean/median/imputation.


```python
# Replace Null Values of Age column with Its Mean
mean_age = round(Titanic_Dataset['Age'].mean())
Titanic_Dataset['Age'].fillna(mean_age, inplace = True)

Titanic_Dataset
```

```python
# Replace Null Values of Cabin Column with unknown
Titanic_Dataset['Cabin'].fillna('unknown', inplace = True)
Titanic_Dataset
```

```python
# Replace Null Values of Embark with most repeatitive value which is 'S'
Embark_mode = Titanic_Dataset['Embarked'].mode() 
Titanic_Dataset['Embarked'].fillna('S', inplace = True)

Titanic_Dataset.isnull().sum()


```

# 3.Convert categorical features into numerical using encoding.


```python
# Encode using mapping
# Common values: S = 0, C = 1, Q = 2
embarked_mapping = {'S': 0, 'C': 1, 'Q': 2}

Titanic_Dataset['Embarked_code'] = Titanic_Dataset['Embarked'].map(embarked_mapping)

Titanic_Dataset
```

# 4.Normalize/standardize the numerical features.


```python
# Min-Max Normalization
Titanic_Dataset['Age_normalized'] = (Titanic_Dataset['Age'] - Titanic_Dataset['Age'].min()) / (Titanic_Dataset['Age'].max() - Titanic_Dataset['Age'].min())
Titanic_Dataset['Fare_normalized'] = (Titanic_Dataset['Fare'] - Titanic_Dataset['Fare'].min()) / (Titanic_Dataset['Fare'].max() - Titanic_Dataset['Fare'].min())

# Z-Score Standardization
Titanic_Dataset['Age_standardized'] = (Titanic_Dataset['Age'] - Titanic_Dataset['Age'].mean()) / Titanic_Dataset['Age'].std()
Titanic_Dataset['Fare_standardized'] = (Titanic_Dataset['Fare'] - Titanic_Dataset['Fare'].mean()) / Titanic_Dataset['Fare'].std()

Titanic_Dataset
```

# 5.Visualize outliers using boxplots and remove them.


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.boxplot(Titanic_Dataset['Age']) 
plt.title('Box Plot for Fare')
plt.ylabel('Fare')
plt.show()
```

```python
#removing outliers

age_data = Titanic_Dataset['Age']

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = age_data.quantile(0.25)
Q3 = age_data.quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
filtered_data = Titanic_Dataset[(Titanic_Dataset['Age'] >= lower_bound) & (Titanic_Dataset['Age'] <= upper_bound)]

# Verify if outliers are removed
print(f'Original dataset size: {Titanic_Dataset.shape[0]}')
print(f'Filtered dataset size (outliers removed): {filtered_data.shape[0]}')
```
